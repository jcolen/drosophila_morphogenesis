```python
import h5py
from morphogenesis.plot_utils import *

with h5py.File('myosin_fitting.h5', 'r') as h5f:
    features = h5f['ensemble/features']
    print(features.keys())

    ncols = 6
    fig, ax = plt.subplots(len(features)//ncols + 1, ncols, dpi=200)
    ax = ax.flatten()
    for i, key in enumerate(features.keys()):
        plot_tensor(ax[i], features[key][25])
        ax[i].set_title(key)
    
    while i < len(ax):
        ax[i].axis('off')
        i += 1

plt.tight_layout()
```
